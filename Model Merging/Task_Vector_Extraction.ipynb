```python
import torch
from transformers import AutoModelForCausalLM

base_model_id = "meta-llama/Meta-Llama-3-8B-Instruct"
fine_tuned_model_id = "MohamedRashad/Arabic-Orpo-Llama-3-8B-Instruct"
#fine_tuned_model_id = "jpacifico/French-Alpaca-Llama3-8B-Instruct-v1.0"
base_model = AutoModelForCausalLM.from_pretrained(base_model_id, quantization_config=bnb_config)
fine_tuned_model = AutoModelForCausalLM.from_pretrained(fine_tuned_model_id, quantization_config=bnb_config)
```

```python
import torch

task_vector = []

for param_base, param_fine_tuned in zip(base_model.parameters(), fine_tuned_model.parameters()):
    diff = param_fine_tuned.data.clone().detach() - param_base.data.clone().detach()
    task_vector.append(diff)

torch.save(task_vector, "task_vector_arabic_alpaca.pt")

print("Task vector saved successfully!")
```

```python
task_vector = torch.load("task_vector_french_alpaca.pt")

new_model = AutoModelForCausalLM.from_pretrained(
    "meta-llama/Meta-Llama-3-8B-Instruct",
    quantization_config=bnb_config
)

for param, delta in zip(new_model.parameters(), task_vector):
    param.data.add_(delta)

new_model.save_pretrained("new_model_with_task_vector")
print("New model created with applied task vector!")
```

```python
import torch
import torch.nn.functional as F
from transformers import AutoModelForCausalLM, AutoTokenizer
from datasets import load_dataset

def compute_perplexity(model, tokenizer, texts, max_samples=900):
    """Compute perplexity over a dataset sample."""
    model.eval()
    total_loss = 0
    count = 0

    for text in texts[:max_samples]:
        if len(text.strip()) == 0:
            continue
        inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512).to("cuda")

        with torch.no_grad():
            outputs = model(**inputs, labels=inputs["input_ids"])
            loss = outputs.loss
            total_loss += loss.item()
            count += 1

    avg_perplexity = torch.exp(torch.tensor(total_loss / count))
    return avg_perplexity.item()
```

```python
from datasets import load_dataset
import sys

dataset = load_dataset("OpenLLM-France/Lucie-Training-Dataset", "AmericanStories", split="train", streaming=True)

total_size = 0
max_size = 1 * 1024 * 1024
filtered_samples = []

for sample in dataset:
    sample_size = sys.getsizeof(str(sample))
    if total_size + sample_size > max_size:
        break
    filtered_samples.append(sample)
    total_size += sample_size

print(f"Downloaded approximately {total_size / (1024)} KB of data")

from datasets import Dataset
dataset = Dataset.from_list(filtered_samples)
text_samples = dataset["text"][:1000]
```

```python
from datasets import load_dataset

# Load Arabic Wikipedia dataset
dataset = load_dataset("wikimedia/wikipedia", "20231101.ar", split="train")

# Extract sentences while filtering empty lines and short sentences (< 3 words)
text_samples = []
for i in range(len(dataset)):
    line = dataset[i]["text"].strip()
    if line and len(line.split()) >= 3:  # Remove empty lines & short lines
        text_samples.append(line)
    if len(text_samples) >= 1000:  # Stop after collecting 1000 valid lines
        break

# Save as a text file (one sentence per line)
output_file = "arabic_1000.txt"

with open(output_file, "w", encoding="utf-8") as f_out:
    for line in text_samples:
        f_out.write(line + "\n")

print(f"Arabic dataset saved as {output_file} with {len(text_samples)} valid lines!")
```
